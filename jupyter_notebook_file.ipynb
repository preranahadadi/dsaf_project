```python
# Assuming 'topic' is the column indicating diabetes (e.g., 'diabetes' for diabetics)
# Filter the DataFrame to include only rows with 'diabetes' in the 'topic' column
diabetic_df = df[df['Topic'] == 'Diabetes']
diabetic_df.reset_index(drop=True, inplace=True)
# Print the DataFrame with rows containing only 'diabetes' in the 'topic' column
display(diabetic_df)
```

```python
diabetic_df.drop('StratificationCategoryID2', axis=1, inplace=True)
diabetic_df.drop('StratificationID2',axis =1 ,inplace=True)
diabetic_df.drop('StratificationCategoryID3', axis=1, inplace=True)
diabetic_df.drop('StratificationID3',axis =1 ,inplace=True)
diabetic_df.drop('Response',axis =1 ,inplace=True)
diabetic_df.drop('StratificationCategory3',axis =1 ,inplace=True)
diabetic_df.drop('Stratification3',axis =1 ,inplace=True)
diabetic_df.drop('StratificationCategory2',axis =1 ,inplace=True)
diabetic_df.drop('Stratification2',axis =1 ,inplace=True)
diabetic_df.drop('ResponseID',axis =1 ,inplace=True)


```
